```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'spacex_dataset.csv')

pd.set_option("display.max_rows", None)

df
```

```python
df['PAYLOAD_MASS__KG_'] = df['PAYLOAD_MASS__KG_'].astype(int)
df['Booster_Version'] = df['Booster_Version'].astype('|S')
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.dtypes
```

### Task 3
#### Display the total payload mass carried by boosters launched by NASA (CRS)

```python
task3 = df[df['Customer'] == "NASA (CRS)"]['PAYLOAD_MASS__KG_'].sum()

task3
```

### Task 6
#### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000

```python
task6 = df[(df['Landing _Outcome'] == 'Success (drone ship)') & (df['PAYLOAD_MASS__KG_'] > 4000) & (df['PAYLOAD_MASS__KG_'] < 6000)]['Booster_Version']

task6
```

### Task 7
#### List the total number of successful and failure mission outcomes

```python
task7_s = df[df['Mission_Outcome'] == 'Success'].value_counts().sum()
task7_f = df[df['Mission_Outcome'].str.contains("Failure")].value_counts().sum()

print("Successful Mission Outcomes: " + str(task7_s))
print("Failed Mission Outcomes: " + str(task7_f))
```

### Task 8
